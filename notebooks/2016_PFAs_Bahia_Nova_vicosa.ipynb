```python
import os
import geopandas

path = os.path.join('data', 'BTS_SHP')
bts = geopandas.read_file(path)
bts = bts['geometry'].values.tolist()[0]
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter

extent = [-39.55, -39.08, -18, -17.60]

projection = ccrs.PlateCarree()
fig, ax = plt.subplots(figsize=(9, 9),
                       subplot_kw=dict(projection=projection))

ax.set_extent(extent)
shp = ax.add_geometries(bts, projection,
                        facecolor='LightGray',
                        edgecolor='black',
                        linewidth=0.25)

ax.set_xticks(np.arange(-39.50, -39, 0.15), crs=projection)
ax.set_yticks(np.arange(-18, -17.5, 0.15), crs=projection)
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()
ax.xaxis.set_major_formatter(lon_formatter)
ax.yaxis.set_major_formatter(lat_formatter)
```
